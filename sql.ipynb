```python
import numpy as np 
import pandas as pd
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
import pandas as pd
import sqlite3
```

```python
## Read the file
df_netflix = pd.read_csv('netflix_titles.csv')
```

```python
## print the columns of all dataframes
print('The columns of the Netflix data frame are :-  ',df_netflix.columns)
```

```python
# import sqlalchemy and create a sqlite engine
from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo=False)

# export the dataframe as a table 'playstore' to the sqlite engine
df_netflix.to_sql("netflix", con =engine)
```

```python
## Test the code by select the complete table
sql='''
Select * from netflix
limit 5

''';
  
df_sql = pd.read_sql_query(sql,con=engine)
df_sql.head()
```

```python
## Query: Min,Max,count,avg,sum

sql='''

Select
min(release_year) as min_release_year,
max(release_year) as max_release_year,
count(distinct release_year) as count_of_release_year,
round(avg(release_year),2) as avg_of_all_release_years,
sum(release_year) as sum_of_all_release_years
from netflix
limit 1

''';
 
df_sql = pd.read_sql_query(sql,con=engine)
df_sql.head()
```

```python
## Query: in

sql='''

select * from netflix
where country in ('India','United States','Australia')
order by country asc

''';
   
df_sql = pd.read_sql_query(sql,con=engine)
df_sql.head()
```

```python
## Query: between

sql='''

select * from netflix
where release_year between 2020 and 2021

''';
    
df_sql = pd.read_sql_query(sql,con=engine)
df_sql.head()
```

```python
## Query: case statements
## Find the total number of data record for three countires: India, United States and South Africa
sql='''

select
sum(case when country='India' then 1 else 0 end) as Shows_in_India,
sum(case when country='United States' then 1 else 0 end) as Shows_in_United_States,
sum(case when country='South Africa' then 1 else 0 end) as Shows_in_South_Africa
from netflix

''';
  
df_sql = pd.read_sql_query(sql,con=engine)
df_sql.head()
```

```python
## Query: Sub Queries
## Create a sub queries to show the title of movie or TV Show only.

sql='''

select distinct(a.title) from (select * from netflix) a

''';
 
df_sql = pd.read_sql_query(sql,con=engine)
df_sql.head()
```

```python
## Query: Row Number
## Check the row numbers which are ordered by title alphabet ascending
sql='''

select 
title,type,director,
ROW_NUMBER() OVER (ORDER BY
        title asc) as row_number
        from netflix
group by title
order by title asc

''';

df_sql = pd.read_sql_query(sql,con=engine)
df_sql.head(10)
```

```python
## Query: coalesce
# syntax: coalesce(column_name,submit value to replace the null value)

sql='''
select coalesce(release_year,0) as first_non_zero_value from netflix
limit 10

''';

df_sql = pd.read_sql_query(sql,con=engine)
df_sql.head()
```
